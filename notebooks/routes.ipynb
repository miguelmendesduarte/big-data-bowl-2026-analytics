```python
import sys
import os

sys.path.append(os.path.abspath("/home/miguelmd/projects/big-data-bowl-2026-analytics"))

from src.io.datasets import CSVReader
from src.core.settings import get_settings

settings = get_settings()
```

```python
plays_data = CSVReader().read(settings.CLEANED_PLAYS_FILE)

print(plays_data.route_of_targeted_receiver.value_counts())
```

```python
PASS_LENGTH_MIN_THRESHOLD = 5
PASS_LENGTH_MAX_THRESHOLD = 200
DROPBACK_TYPE = ["TRADITIONAL"]
ROUTE = ["OUT", "IN", "HITCH"]
PASS_RESULT = ["C", "I", "IN"]  # C: Complete pass, I: Incomplete pass, S: Quarterback sack, IN: Intercepted pass, R: Scramble
```

```python
plays_data = plays_data[
    (plays_data["route_of_targeted_receiver"].isin(ROUTE))
    & (plays_data["pass_length"] > PASS_LENGTH_MIN_THRESHOLD)
    & (plays_data["pass_length"] < PASS_LENGTH_MAX_THRESHOLD)
    & (plays_data["dropback_type"].isin(DROPBACK_TYPE))
    & (plays_data["pass_result"].isin(PASS_RESULT))
]

print(f"Found {len(plays_data)} plays with the following criteria:")
print(f"  - Route(s): {', '.join(ROUTE)}")
print(f"  - Pass Length: {PASS_LENGTH_MIN_THRESHOLD} to {PASS_LENGTH_MAX_THRESHOLD} yards")
print(f"  - Dropback Type(s): {', '.join(DROPBACK_TYPE)}")
print(f"  - Pass Result(s): {', '.join(PASS_RESULT)}")
```
